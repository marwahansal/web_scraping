```python
import requests as rq
from bs4 import BeautifulSoup as bs
req = rq.get('https://pokemondb.net/pokedex/national')
status= req.status_code
print(status)
encoding = req.encoding
html_doc = req.text
soup = bs (html_doc,"html.parser")
name=soup.find_all("a", {"class":"ent-name"})
print(name)
import csv 
images=[]
for link in range(len(images)):
    l=images[link]
    print(l)
def get_img(url_img,id):
    r = rq.get(url_img)
    if r.status_code == 200:
        img_container="C:\\Users\\dell\\Desktop\\Pokemon_Scraping\\images"+str(id)+".jpg"
        with open(img_container, 'wb') as f:
            f.write(r.content)

for img in name:
    images.append(img["href"])
print(images)

id=1
for link in range(len(images)):
        pk_url = 'https://pokemondb.net'+ images[link]
        req_pk = rq.get(pk_url)
        html_doc_pk = req_pk.text
        soup_pk = bs(html_doc_pk, 'html.parser')
        img_src=soup_pk.find("img").get("src").encode('utf-8')
        get_img(img_src,id)
        id=id+1
```
