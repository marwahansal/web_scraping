# Pokemon(ID,NAME)

```python
import requests
import bs4 as bs
```

```python
req=requests.get("https://pokemondb.net/pokedex/national?fbclid=IwAR3XHXrBQQ9RjiAinVxDETVQkn9CdtRSMnx6Qgk4Sn96zOhuA3450e37rUk")
```

```python
soup=bs.BeautifulSoup(req.text,'html.parser')
```

```python
data=soup.find_all('div',{'class':'infocard'})
```

```python
import pandas as pd
df=pd.DataFrame(columns={'ID','Nom'})
for item in data:  
  Nom=item.find_all('a',{'class':'ent-name'})[0].text.encode("utf_8")
  id=item.find_all('small')[0].text.encode("utf_8")
  
  df=df.append({'ID':id.decode("utf-8"),'Nom':Nom.decode("utf-8")},ignore_index=True)
print(df)
```
